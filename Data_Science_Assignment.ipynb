# **Lahore Faculty**




```python
URL = "https://lhr.nu.edu.pk/faculty/"
res = requests.get(URL)

print(res)
```

```python
soup = BeautifulSoup(res.content,"html.parser")

print(soup)
```

```python
all_dept = soup.find_all(lambda tag: tag.name == 'div' and tag.get('class') == ['container'], id=True )
print(len(all_dept))
print(all_dept)
```

```python
df = pd.DataFrame(columns=['ID','Name','Designation','HEC Approved Supervisor', 'Highest Education','Email','Department','Extension','Image Url'])
print(df.head())
```

```python
for dept in all_dept:
  dept_name = dept.find('h1',class_ = "mb-2 mt-3").text.strip()
  print(dept_name)

  all_faculty = dept.find('div',class_ = "container tab-pane active").find_all('div',class_ = "col-lg-3 col-md-4 col-sm-6 col-12")
  print(len(all_faculty))

  for faculty in all_faculty:
      profile_link = faculty.find('a').get('href')

      profile_url = "https://lhr.nu.edu.pk" + profile_link

      profile_res = requests.get(profile_url) # Requesting the Profile Page of a Faculty Member

      profile_soup = BeautifulSoup(profile_res.content,"html.parser")

      profile = profile_soup.find('div',class_="row")

      link_split = profile_link.split("/")

      faculty_id = link_split[len(link_split)-1] ## Faculty Id Extraction from the Profile Link

      faculty_img = profile.find('img').get('src')

      info = profile.find('div', class_= "card-body pr-0 pl-0 pb-0")

      name = info.find('h5',class_ = "text-center").text.strip()

      designation_info = info.find('p',class_ = "small text-center font-italic").text.strip().split("\n")
      designation = designation_info[0] + " " + (designation_info[1].strip()) if len(designation_info) > 1 else designation_info[0]
      approved = True if len(designation_info) == 4 and designation_info[3].strip() == "HEC Approved PhD Supervisor" else False

      email = info.find('p',class_ = "mb-0 text-center").text.strip()
      extension = info.find('p', class_ = "mt-0 text-center").find('span').text.strip().split(" ")[-1].split(":")[-1]


      edu = profile.find('div',class_ = "col-lg-8 col-md-6 col-sm-12 text-justify").find_all('li')

      if edu:
        edu = edu[0].text.strip()
      else:
        edu = None

      print(faculty_id,name,designation,approved,faculty_img,email,extension,edu)

      facultyDf = pd.DataFrame([{'ID':faculty_id,'Name':name,'Designation':designation,'HEC Approved Supervisor':approved,'Highest Education':edu,'Email':email,'Department':dept_name,'Extension':extension,'Image Url':faculty_img}])

      df = pd.concat([df,facultyDf],ignore_index=True)


print(df.head())

```

# **Islamabad Faculty**

```python
URL = "http://isb.nu.edu.pk/Faculty/allfaculty/"
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:135.0) Gecko/20100101 Firefox/135.0"
}

res = requests.get(URL,headers=headers);
print(res)

```

```python
soup_2 = BeautifulSoup(res.content,"html.parser")

print(soup_2.prettify())

depts = soup_2.find(class_ = "content-block").findAll(lambda tag: tag.name == 'div' and tag.get('class') == ['container'])

print((depts))
```

```python
df = pd.DataFrame(columns=['ID','Name','Designation','HEC Approved Supervisor', 'Highest Education','Email','Department','Extension','Image Url'])
print(df.head())
```

```python
ids = [301,302,303,313]

for dept in depts:
  dept_name = dept.find('h3').find(class_ = "animated fadeInLeft").text.strip()
  print(dept_name)

  ## As the Faculty is loaded Dynamically using Ajax so we will execute a POST Request to obtain the Employees Record

  headers = {

    "Host": "isb.nu.edu.pk",
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:135.0) Gecko/20100101 Firefox/135.0",
    "Accept": "*/*",
    "Origin": "http://isb.nu.edu.pk",
    "Connection": "keep-alive",
    "Referer": "http://isb.nu.edu.pk/Faculty/allfaculty/",
    "Content-Length": "0",
  }

  response = requests.post("http://isb.nu.edu.pk/Faculty/GetAllEmp?id="+str(ids.pop(0)), headers=headers)

  if response.status_code == 200:
    all_faculty = response.json()


    for faculty in all_faculty:
      faculty_id = faculty['Emp_ID']
      name = faculty['Name'].strip()
      email = faculty['Email'].strip()
      approved = False if faculty['hec_supervisor'] == None else faculty['hec_supervisor']
      extension = faculty['Extension']
      designation = faculty['Designation_Title'].strip()
      faculty_img = faculty['ImagePath']


      details_res = requests.post("http://isb.nu.edu.pk/Faculty/Details1",data={"id": faculty_id} , headers=headers)

      edu = "Not Specified"

      if details_res.status_code == 200:
        list_edu = details_res.json()['listEEdu']
        edu = list_edu.pop(0)['Degree_Name'].strip() if len(list_edu) > 0 else "Not Specified"

      #print(faculty_id,name,designation,approved,faculty_img,email,extension,edu)
      facultyDf = pd.DataFrame([{'ID':faculty_id,'Name':name,'Designation':designation,'HEC Approved Supervisor':approved,'Highest Education':str(edu),'Email':email,'Department':dept_name,'Extension':extension if extension else "None",'Image Url':faculty_img}])

      df = pd.concat([df,facultyDf],ignore_index=True)


df.to_csv('isb.csv')
print(df.head())

```

# **Faislabad Faculty**

```python
depts = {"Department of Computer Science" : "https://cfd.nu.edu.pk/department-cs/",
        "Department of Software Engineering": "https://cfd.nu.edu.pk/department-se/",
        "Department of Artificial Intelligence": "https://cfd.nu.edu.pk/department-ai/",
        "Department of Electrical Engineering": "https://cfd.nu.edu.pk/department-ee/",
        "FAST School of Management": "https://cfd.nu.edu.pk/department-fsm/",
        "Department of Sciences & Humanities": "https://cfd.nu.edu.pk/department-sh/",
        }
```

```python
df = pd.DataFrame(columns=['ID','Name','Designation','HEC Approved Supervisor', 'Highest Education','Email','Department','Extension','Image Url'])
print(df.head())
```

```python
itr = 0
for key,value in depts.items():
  URL = value;
  dept_name = key;

  headers = {
      "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:135.0) Gecko/20100101 Firefox/135.0"
  }

  cfd_res = requests.get(URL,headers=headers)

  print(cfd_res)

  cfd_soup = BeautifulSoup(cfd_res.content,"html.parser")

  all_faculty = cfd_soup.find('div',class_ = "unitech-teachers").find_all('div',class_ = "col-md-3 col-sm-6 col-xs-12")

  for faculty in all_faculty:
    link = faculty.find('a').get('href') ## MORE LINK


    ## IMAGE
    faculty_img = faculty.find('img').get('src')

    #NAME
    name = faculty.find('h4').find('a').text.strip()


    # DESIGNATION
    designation = faculty.find('h6').text.strip()

    ## OTHER INFOS
    info = faculty.find('div',class_ = "unitech-teacher__hover__info")
    email = info.find('li').find('p').text.strip()
    hec = info.find('li').find('p',class_ = "hec")
    approved = True if hec is not None else False

    ## REQUESTING INDIVIDUAL FACULTY INFO

    faculty_res = requests.get(link,headers=headers)

    faculty_soup = BeautifulSoup(faculty_res.content,"html.parser")

    address = faculty_soup.find('ul',class_ ="teacher__address").find_all('li')

    extension = address[3].text.split(":")

    extension = extension[1].strip() if len(extension) > 1 else "None"

    edu = faculty_soup.find('div',class_ = "htc__skill__container progress__bar--2").find('p')

    if edu is None:
      edu = faculty_soup.find('div',class_ = "htc__skill__container progress__bar--2").find('ul')

      if edu is None:
        edu = "Not Specified"
      else:
        edu = edu.find('li').find('strong')

      if edu is None:
        edu = faculty_soup.find('div',class_ = "htc__skill__container progress__bar--2").find('ul')

        if edu is None:
          edu = "Not Specified"
        else:
          edu = edu.find('li')

        if edu is None:
          edu = "Not Specified"

    if edu is not None and edu != "Not Specified" :
      edu = edu.text.split("\n")[0].strip()


    print(name,designation,email,approved,extension,edu,key)
    itr+=1
    facultyDf = pd.DataFrame([{'ID':str(itr) ,'Name':name,'Designation':designation.strip(),'HEC Approved Supervisor':approved,'Highest Education':str(edu).strip(),'Email':email,'Department':key,'Extension':extension if extension else "None",'Image Url':faculty_img}])

    df = pd.concat([df,facultyDf],ignore_index=True)


df.to_csv('cfd.csv')
print(df.head())

```

# **Peshawer Faculty**

```python
depts = {"Department of Computer Science" : "http://pwr.nu.edu.pk/cs-faculty/",
        "Department of Electrical Engineering": "http://pwr.nu.edu.pk/ee-faculty/",
        "Department of Sciences & Humanities": "http://pwr.nu.edu.pk/sh-faculty/",
        }
```

```python
df = pd.DataFrame(columns=['ID','Name','Designation','HEC Approved Supervisor', 'Highest Education','Email','Department','Extension','Image Url'])
print(df.head())
```

```python
for key,value in depts.items():
  URL = value;
  dept_name = key;

  res = requests.get(URL)
  print(res)

  khan_soup = BeautifulSoup(res.content,"html.parser")

  all_faculty = khan_soup.find_all("div",class_ = "faculty-member")


  ## ITERATING OVER THE FACULTIES PER DEPT

  for faculty in all_faculty:
      faculty_img = faculty.find('img',class_ = "faculty-photo").get('src')

      details = faculty.find('div',class_ = "faculty-details")
      h2 = details.find('h2')
      a = h2.find('a')
      link = a.get('href')

      name = a.text.strip()

      faculty_id = link.split("=")[1]

      info = details.find_all('p')

      designation = info[0].text.strip()
      approved = False
      if info[1].text.strip() == "HEC/University Approved PhD Supervisor":
        approved = True
        email = info[2].text.strip()
      else:
        email = info[1].text.strip()


      extension = info[-1].text.split(" ")[-1].strip(")").strip()

      extension = extension if len(extension) != 0  else "None"

      single_res = requests.get("http://pwr.nu.edu.pk/faculty-profile/?id="+faculty_id)
      single_soup = BeautifulSoup(single_res.content,"html.parser")

      edu_info = single_soup.find_all(lambda tag: tag.name == 'div' and tag.get('class') == ['container'])[2].find("div",class_ = "section-content").find("div",class_ = "col-md-12")
      edu_list = edu_info.find('ul')
      edu = edu_list.find('li').text.strip()


      facultyDf = pd.DataFrame([{'ID':faculty_id ,'Name':name,'Designation':designation.strip(),'HEC Approved Supervisor':approved,'Highest Education':str(edu).strip(),'Email':email,'Department':key,'Extension':extension if extension else "None",'Image Url':faculty_img}])

      df = pd.concat([df,facultyDf],ignore_index=True)


df.to_csv('pwr.csv')
print(df.head())

```

# **MERGING ALL FACULTIES**

```python
## Concatenating all CSV into 1

lhr_df = pd.read_csv("lhr.csv")
isb_df = pd.read_csv('isb.csv')
khi_df = pd.read_csv("khi.csv")
cfd_df = pd.read_csv("cfd.csv")
pwr_df = pd.read_csv("pwr.csv")

faculty_df = pd.concat([lhr_df,isb_df,khi_df,cfd_df,pwr_df],ignore_index=True)
faculty_df = faculty_df.drop(columns=["Unnamed: 0"])

faculty_df['Extension'] = pd.to_numeric(faculty_df['Extension'], errors='coerce').astype('Int64')
print(faculty_df.dtypes)

```
